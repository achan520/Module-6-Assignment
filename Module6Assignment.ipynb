```python
import csv
import pandas as pd
```

```python
# Read csv file
file = 'IMDb_All_Genres_etf_clean1.csv'
df = pd.read_csv(file)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Select relevant feature and target variable
feature = 'Runtime(Mins)'
target = 'Rating'
```

```python
# Drop rows with missing values in the target variable and the feature
df.dropna(subset=[target, feature], inplace=True)
```

```python
# Extract feature (X) and target variable (y)
X = df[feature]]
y = df[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
predicted_ratings = model.predict(X_test)

# Retrieve the actual ratings from the test set
actual_ratings = y_test.values
movie_titles = df.loc[X_test.index, 'Movie_Title'].values

# Print the first 5 samples with their predicted and actual ratings and their movie title
for i in range(5):
    print("Movie Title:", movie_titles[i])
    print("Predicted Rating:", predicted_ratings[i])
    print("Actual Rating:", actual_ratings[i])
    print()
```
